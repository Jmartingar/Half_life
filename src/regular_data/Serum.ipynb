```python
import pandas as pd
import importlib.util
import os

funct_path= os.path.abspath('../../utils/functions.py')
```

```python
# Importa funciones desde archivo externo
spec = importlib.util.spec_from_file_location("functions", funct_path)
funct = importlib.util.module_from_spec(spec)
spec.loader.exec_module(funct)
```

```python
df = pd.read_excel('../../data/raw/Estimating_Serum.xlsx', header=1)
```

```python
df=df.rename(columns={'Sequence':'sequence', 'Half-life (hours)': 'half_life_seconds', 'Modifications':'modifications', 'Assay': 'experimental_characteristics'})
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df['modifications'] = df['modifications'].replace('N.A.', '')
funct.mod_torf(df)
df['sequence'] = df['sequence'].str.upper()
```

```python
df['hl_is_float'] = df['half_life_seconds'].apply(funct.float_or_false)
df_float=df[df['hl_is_float']==True]
df_object=df[df['hl_is_float']==False]
df_float.reset_index(drop=True, inplace=True)
df_object.reset_index(drop=True, inplace=True)
```

```python
for i in range (len(df_float)):
    df_float.loc[i,'half_life_seconds'] = float(df_float.loc[i,'half_life_seconds'])*3600
```
