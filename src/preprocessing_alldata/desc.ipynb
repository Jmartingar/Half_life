```python
import pandas as pd
```

```python
antiviral=pd.read_excel('../../data/raw/antiviral_peptides.xlsx', header=0)
hlp_10=pd.read_csv("../../data/regular/HLP_10.csv")
hlp_16=pd.read_csv("../../data/regular/HLP_16.csv")
peplife=pd.read_csv("../../data/regular/PEP_LIFE.csv")
plifepred=pd.read_csv("../../data/regular/Plifepred.csv")
serum=pd.read_csv("../../data/regular/Serum.csv")
```

```python
hlp=pd.concat([hlp_10, hlp_16], ignore_index=True)
```

```python
dataframes = {
    "Antiviral": antiviral,
    "HLP": hlp,
    "PEP_LIFE": peplife,
    "Plifepred": plifepred,
    "Serum": serum
}
```

```python
for name, df in dataframes.items():
    print(f"{name}: {df.shape[0]} filas, {df.shape[1]} columnas")
```

```python
import matplotlib.pyplot as plt

# Obtener nombres y tamaños (número de filas)
nombres = list(dataframes.keys())
tamanos = [df.shape[0] for df in dataframes.values()]

# Crear gráfico de barras
plt.figure(figsize=(6, 3))
plt.barh(nombres, tamanos, color='skyblue')
plt.ylabel('Dataset')
plt.xlabel('Número de filas')
plt.title('Tamaño de cada DataFrame')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../../img/datasets/dataset_sizes_raw.png', dpi=300)
plt.show()
```
