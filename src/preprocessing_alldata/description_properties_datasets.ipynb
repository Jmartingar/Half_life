```python
import pandas as pd
import sys
sys.path.insert(0, "../../")
from utils.PeptidesDescriptors import PeptidesDescriptors
import os
folder = '../../data/canon/'
```

```python
dataframes = {}
for archive in os.listdir(folder):
    if archive.endswith('.csv') and os.path.isfile(os.path.join(folder, archive)):
        df = pd.read_csv(os.path.join(folder, archive))
        df.reset_index(inplace=True)
        df = df[['sequence', 'hl_category','half_life_seconds', 'length_sequence']]
        df['dataset_name'] = archive
        dataframes[archive] = df
```

```python
df_concat = pd.concat(dataframes.values(), ignore_index=True)
```

```python
peptide_descriptor_instance = PeptidesDescriptors(
    dataset=df_concat,
    column_seq="sequence"
)

peptide_descriptor_instance.apply_physicochemical_properties()
df=peptide_descriptor_instance.dataset
```
