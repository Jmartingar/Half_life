Test de normalidad para saber si los datos tienen una distribucion normal

Conclusion no siguen una distribucion normal

```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
from scipy.stats import shapiro
from scipy.stats import anderson
```

```python
df=pd.read_csv('../../data/described/described_antiviral_homology_90.csv')
```

```python
column_names = [
    'Molecular Weight', 'Isoelectric point', 'Charge density', 'Charge',
    'Instability index', 'Aromaticity', 'Aliphatic index', 'Boman index', 'Hydrophobic ratio',
    'freq_A', 'freq_C', 'freq_D', 'freq_E', 'freq_F', 'freq_G', 'freq_H', 'freq_I',
    'freq_N', 'freq_K', 'freq_L', 'freq_M', 'freq_P', 'freq_Q', 'freq_R',
    'freq_S', 'freq_T', 'freq_V', 'freq_W', 'freq_Y'
]
```

```python
for col in column_names:
    result = anderson(df[col])
    print(f'Estadística de la prueba: {result.statistic}, valores críticos: {result.critical_values} para columna: {col}')
```

```python
for col in column_names:
    stat, p_value = shapiro(df[col])
    #print(f'Estadística de la prueba: {stat}, p-valor: {p_value}')
    if p_value > 0.05:
        print(f'La variable {col} sigue una distribución normal')
    else:
        print(f'La variable {col} no sigue una distribución normal')
```

```python
for col in column_names:
    stats.probplot(df[col], dist="norm", plot=plt)
    plt.show()
```
